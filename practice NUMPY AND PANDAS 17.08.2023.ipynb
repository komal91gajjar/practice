# 	String Functions

```python
a = np.array(["hello"])
b = np.array(["python"])
str_concate = np.char.add(a, b)
str_concate
```

```python
a = np.array(["Hello", "python"]) # string concate
b = np.array([" ", " !"])
print(np.char.add(a,b))
```

```python
print(np.char.multiply(a , 2)) # string multiply
```

```python
print(np.char.str_len(a)) # string find lengh
```

```python
a = np.array(["apple","banana","cherry"])
b = np.array(["watermelon", "banana","apple"])

result = np.char.not_equal(a,b)
result
```

```python
a = np.array(["apple","banana","cherry"])
b = np.array(["watermelon", "banana","apple"])

result = np.char.equal(a,b)
result
```

```python
a = np.array(["Hello", "Python", "World"])

result = np.char.lower(a)
print(result)

result1 = np.char.upper(a)
print(result1)
```
