```python
batsman = pd.read_csv("D:\\excel data\\sharma-kohli.csv")
```

```python
batsman
```

```python
plt.plot(batsman["index"],batsman["V Kohli"])
```

```python
# how to compare one or more columns in plotting

plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])
plt.title("Rohit sharma v/s Virat kohli")
plt.xlabel("year")
plt.ylabel("Run score")
```

```python
# how to change a colors of default colors, linestyle,marker

plt.plot(batsman["index"],batsman["V Kohli"], color = "green",linestyle = "solid",linewidth = 3.5,marker = "D",markersize = 10)
plt.plot(batsman["index"],batsman["RG Sharma"], color = "blue",linestyle = "dotted",linewidth = 2.5,marker = "*",markersize = 15)
plt.title("Rohit sharma v/s Virat kohli")
plt.xlabel("year")
plt.ylabel("Run score")
```

```python
# how to add label in graph

plt.plot(batsman["index"],batsman["V Kohli"], color = "green",linestyle = "solid",linewidth = 3.5,marker = "D",markersize = 10,label = "virat")
plt.plot(batsman["index"],batsman["RG Sharma"], color = "blue",linestyle = "dotted",linewidth = 2.5,marker = "*",markersize = 15,label = "Rohit")
plt.title("Rohit sharma v/s Virat kohli")
plt.xlabel("year")
plt.ylabel("Run score")
plt.legend()
```

```python
plt.plot(batsman["index"],batsman["V Kohli"], color = "green",linestyle = "solid",linewidth = 3.5,marker = "D",markersize = 10,label = "virat")
plt.plot(batsman["index"],batsman["RG Sharma"], color = "blue",linestyle = "dotted",linewidth = 2.5,marker = "*",markersize = 15,label = "Rohit")
plt.title("Rohit sharma v/s Virat kohli")
plt.xlabel("year")
plt.ylabel("Run score")
plt.legend()
plt.grid()
plt.show()
```

# bar chart

```python
#Bivariant analysis
# Numerical column v/s Categorical column
# Use case: Aggregate analysis of group
```

```python
df = pd.read_csv("D:\\excel data\\batsman_season_record.csv")
```

```python
df
```

```python
plt.bar(df["batsman"],df["2015"])
```

```python
# how to plot multipal bar graph

plt.bar(np.arange(df.shape[0])-0.2,df["2015"],width=0.2)
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2)
plt.xticks(np.arange(df.shape[0]),df["batsman"])
plt.show()
```

# stack bar chart

```python
df
```

```python
plt.bar(df["batsman"],df["2017"],label = "2017")
plt.bar(df["batsman"],df["2016"],bottom=df["2017"],label = "2016")
plt.bar(df["batsman"],df["2015"],bottom=(df["2016"] + df["2017"]),label = "2015")
plt.legend()
plt.show()
```

# pie chart

```python
# univariate and bivariate analysis
# Categoricals v/s Numerical
# Use case : to find continue on standard scale
```

```python
# simple pie chart

data = [23,45,100,20,45]
subject = ["english","hindi","maths","gujarati","marathi"]
plt.pie(data, labels = subject)
plt.show()
```

```python
ipl = pd.read_csv("D:\\excel data\\gayle-175.csv")
```

```python
ipl
```

```python
plt.pie(ipl["batsman_runs"],labels=ipl["batsman"],autopct="%0.1f%%")
plt.show()
```

```python
# Explode

plt.pie(ipl["batsman_runs"],labels=ipl["batsman"],autopct="%0.1f%%",explode=[0.1,0.1,0.1,0.1,0,0.1])
plt.show()
```

```python
# shadow

plt.pie(ipl["batsman_runs"],labels=ipl["batsman"],autopct="%0.1f%%",explode=[0.1,0.1,0.1,0.1,0,0.1],shadow=True)
plt.show()
```
