```python
ipl_df["all_players"] = ipl_df["Team1Players"] + ipl_df["Team2Players"]
ipl_df["all_players"]
ipl_df.head(2)
```

```python
def Did_kohli_play(players_list):
    return "V Kohli" in players_list

ipl_df["did_kohli_play"] = ipl_df["all_players"].apply(Did_kohli_play)
```

```python
ipl_df.head(2)
```

```python
# how to use drop_duplicates()

ipl_df[(ipl_df["City"] == "Delhi") & (ipl_df["did_kohli_play"] == True)].drop_duplicates(subset=["City","did_kohli_play"],keep="first")
```
