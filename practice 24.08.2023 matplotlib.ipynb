```python
price1 = df["price1"]
actprice1 = df["actprice1"]

def profit_loss(price1,actprice1):
    if actprice1 > price1:
        return "profit"
    else:
        return "loss"
        
df["result"] = df.apply(lambda row: profit_loss(row["price1"], row["actprice1"]), axis=1)

df.to_csv("D:\\excel data\\trainflipkart.csv", index=False)
        
        
```
