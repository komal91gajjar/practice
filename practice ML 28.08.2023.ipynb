```python
import numpy as np
import matplotlib.pyplot as plt


mean = [0, 0]  # Mean of the distribution
cov = [[1, 0.5], [0.5, 1]]  # Covariance matrix
num_points = 50

data = np.random.multivariate_normal(mean,cov,num_points)

# Create a scatter plot
plt.scatter(data[:, 0], data[:, 1])
plt.xlabel('X',size = 15,color = "r")
plt.ylabel('Y',size = 15, color = "r")
plt.title('Scatter Plot of Bivariate Normal Distribution',color = "r")
plt.grid(True,color = "b",linestyle = "--")
plt.show()
```
