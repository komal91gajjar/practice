```python
matches = pd.read_csv("D:\\excel data\\matches.csv")
matches.head(2)
```

```python
delivery = pd.read_csv("D:\\excel data\\deliveries.csv")
delivery.head(2)
```

```python
ipl_data = delivery.merge(matches,left_on="match_id",right_on="id")
ipl_data.sample()
```

```python
top_50 = ipl_data.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False).head(50).index.tolist()
```

```python
new_ipl = ipl_data[ipl_data["batsman"].isin(top_50)]
```

```python
runs = new_ipl.groupby("batsman")["batsman_runs"].sum()
```

```python
balls = new_ipl.groupby("batsman")["batsman_runs"].count()
```

```python
strike_rate = (runs/balls)*100
```

```python
strike_rate = strike_rate.reset_index()
```

```python
strike_rate
```

```python
out = ipl_data[ipl_data["player_dismissed"].isin(top_50)]
```

```python
not_out = ipl_data["player_dismissed"].value_counts()
```

```python
avg = runs/not_out
```

```python
avg = avg.reset_index()
```

```python
avg.rename(columns={"index":"batsman",0:"avg"},inplace=True)
```

```python
avg = avg.merge(strike_rate,on="batsman")
```

```python
avg
```

# scatter

```python
# plot a scatter plot
trace = go.Scatter(x = avg["avg"],y = avg["batsman_runs"], mode="markers",text=avg["batsman"],marker={"color":"red", "size":12})
data = [trace]
layout = go.Layout(title="batsman average run vs strike rate",xaxis={"title":"batsman average run"}, yaxis={"title":"batsman strike rate"})
# layout = go.Layout(title="batsman run vs strike rate",xaxis={"title":"batsman run"},yaxis={"title":"strike_rate"})
fig = go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

```python
new_ipl_data= ipl_data[["batsman","batsman_runs","inning"]]
```

```python
iw = top_10df.groupby(["batsman","inning"])["batsman_runs"].sum().reset_index()
mask = iw["inning"]==1
mask1 = iw["inning"]==2
one = iw[mask]
two = iw[mask1]

one.rename(columns={"batsman_runs":"1st_inning"},inplace=True)
two.rename(columns={"batsman_runs":"2nd_inning"},inplace=True)

final = one.merge(two,on="batsman")[["batsman","1st_inning","2nd_inning"]]
final
```

```python
# stacked bar graph
trace = go.Bar(x = final["batsman"], y = final["1st_inning"],name = "1st_inning")
trace1 = go.Bar(x = final["batsman"], y = final["2nd_inning"],name = "2nd_inning")
data = [trace,trace1]
layout = go.Layout(title="top 10 ipl batsman",xaxis={"title":"batsman"},yaxis={"title":"runs"},barmode="stack")

fig = go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

```python
# overlayed bar graph
trace = go.Bar(x = final["batsman"], y = final["1st_inning"],name = "1st_inning")
trace1 = go.Bar(x = final["batsman"], y = final["2nd_inning"],name = "2nd_inning")
data = [trace,trace1]
layout = go.Layout(title="top 10 ipl batsman",xaxis={"title":"batsman"},yaxis={"title":"runs"},barmode="overlay")

fig = go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

```python
#nested bar graph
trace = go.Bar(x = final["batsman"], y = final["1st_inning"],name = "1st_inning")
trace1 = go.Bar(x = final["batsman"], y = final["2nd_inning"],name = "2nd_inning")
data = [trace,trace1]
layout = go.Layout(title="top 10 ipl batsman",xaxis={"title":"batsman"},yaxis={"title":"runs"})

fig = go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

# box plot

```python
match_agg = delivery.groupby(["match_id"])["total_runs"].sum().reset_index()
season_vise = match_agg.merge(matches,left_on="match_id",right_on="id")[["match_id","total_runs","season"]]
season_vise
```

```python
trace = go.Box(x = season_vise["total_runs"],name = "all season")
data = [trace]
layout = go.Layout(title = "total score analysis",xaxis={"title":"total score"})
fig = go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

```python
trace = go.Box(x = season_vise[season_vise["season"] == 2017]["total_runs"],name = "2017",marker={"color":"red"})
trace1 = go.Box(x = season_vise[season_vise["season"] == 2008]["total_runs"],name = "2008",marker={"color":"blue"})
data = [trace,trace1]
layout = go.Layout(title = "total score analysis",xaxis={"title":"total score"})
fig = go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

# histogram

```python
x = delivery.groupby("batsman")["batsman_runs"].count()>150
x = x[x].index.tolist()
```

```python
new = delivery[delivery["batsman"].isin(x)]

runs = new.groupby("batsman")["batsman_runs"].sum()
balls = new.groupby("batsman")["batsman_runs"].count()

strike_rate = (runs/balls)*100
strike_rate = strike_rate.reset_index()
strike_rate

```

```python
trace=go.Histogram(x =strike_rate["batsman_runs"],xbins={"size":2})
data = [trace]
layout = go.Layout(title = "strike rate variation",xaxis={"title":"batsman_runs"})
fig = go.Figure(data = data,layout = layout)
pyo.plot(fig)
```
